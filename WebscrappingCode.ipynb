```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from random import randint
from datetime import datetime
driver = webdriver.Chrome()
driver.maximize_window()

def load_more():
    try:
        load_more_button = driver.find_element(By.CLASS_NAME, 'ULvh-button show-more-button')
        load_more_button.click()
        time.sleep(10)
    except Exception as e:
        print("Load more button not found:", e)

def start_kayak(city_from, city_to, date_start,NAME):
    """City codes - it's the IATA codes!
    Date format -  YYYY-MM-DD"""
    kayak = f'https://www.kayak.co.in/flights/{city_from}-{city_to}/{date_start}?sort=bestflight_a'
    driver.get(kayak)
    time.sleep(20)
    load_more()

    # Scraping flight prices
    try:
        flight_elements = driver.find_elements(By.CLASS_NAME, "f8F1-price-text-container")
        flight_prices = [elem.text for elem in flight_elements]
        flight_companies = driver.find_elements(By.CLASS_NAME, "J0g6-operator-text")
        flight_company = [elem.text for elem in flight_companies]
        flight_info = driver.find_elements(By.CLASS_NAME, "JWEO-stops-text")
        flight_type = [elem.text for elem in flight_info]
        flight_durations = driver.find_elements(By.CLASS_NAME, "xdW8-mod-full-airport")
        flight_duration = [elem.text for elem in flight_durations]
       
        return pd.DataFrame({
            "Company": flight_company,
            "Price": flight_prices,
            "Type": flight_type,
            "Duration": flight_duration,
            "City From": city_from,
            "City To": city_to,
            "Date": date_start,
            "name_OF_CITY": NAME,
        })
        
    except Exception as e:
        print("Search Error:", e)
        return pd.DataFrame()

# Load Excel file into a pandas DataFrame
df = pd.read_excel('sin.xlsx')

# Create an empty DataFrame to store all results
all_flights_df = pd.DataFrame()

for index, row in df.iterrows():
    city_name = row['NAME']
    iata_code = row['IATA']
    date = row['DATE']
    flight_data = start_kayak('DEL', iata_code, date, city_name)
    all_flights_df = pd.concat([all_flights_df, flight_data], ignore_index=True)

# Save all data to a single Excel file
all_flights_df.to_excel('outputfile.xlsx', index=False)
driver.quit()
```
